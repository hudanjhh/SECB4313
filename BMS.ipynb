```python
dna = pd.read_csv(data_path + 'BRCA_DNAm.csv')
mrna = pd.read_csv(data_path + 'BRCA_mRNA.csv')
mirna = pd.read_csv(data_path + 'BRCA_miRNA.csv')
label = pd.read_csv(data_path + 'BRCA_label.csv')
```

## **Pre-Processing**

```python
# Check for missing values
dna_missing_value = dna.isnull().sum().sum()
mrna_missing_value = mrna.isnull().sum().sum()
mirna_missing_value = mirna.isnull().sum().sum()

print('DNA missing values:', dna_missing_value)
print('mRNA missing values:', mrna_missing_value)
print('miRNA missing values:', mirna_missing_value)

# Check for duplicate number
print('\nDNA duplicates number: ', dna.duplicated().sum())
print('mRNA Methylation duplicates number: ', mrna.duplicated().sum())
print('miRNA duplicates number: ', mirna.duplicated().sum())
```

```python
# Drop the feature column
dnadrop = dna.drop(['Unnamed: 0'], axis=1)
mrnadrop = mrna.drop(['Unnamed: 0'], axis=1)
mirnadrop = mirna.drop(['Unnamed: 0'], axis=1)
```

```python
dnadrop.head()
```

```python
# Data Transposition
dnaT = dnadrop.transpose()
mrnaT = mrnadrop.transpose()
mirnaT = mirnadrop.transpose()
print('DNA \n', dnaT.shape)
print('mRNA \n', mrnaT.shape)
print('miRNA \n', mirnaT.shape)
```

```python
dnaT.head()
```

```python
dnaT_change = dnaT.reset_index(drop=False)
mrnaT_change = mrnaT.reset_index(drop=False)
mirnaT_change = mirnaT.reset_index(drop=False)
```

```python
dnaT_change.head()
```

```python
print(dnaT_change.shape)
print(mrnaT_change.shape)
print(mirnaT_change.shape)
```

```python
dnaT_change.rename(columns={'index': 'sample ID'}, inplace=True)
mrnaT_change.rename(columns={'index': 'sample ID'}, inplace=True)
mirnaT_change.rename(columns={'index': 'sample ID'}, inplace=True)
```

```python
dnaT_change.head()
```

```python
# Check for common ID
common_ids = set(dnaT_change['sample ID']).intersection(mrnaT_change['sample ID']).intersection(mirnaT_change['sample ID'])
```

```python
print(common_ids)
print(len(common_ids))
```

```python
dna_filtered = dnaT_change[dnaT_change['sample ID'].isin(common_ids)]
mrna_filtered = mrnaT_change[mrnaT_change['sample ID'].isin(common_ids)]
mirna_filtered = mirnaT_change[mirnaT_change['sample ID'].isin(common_ids)]
```

```python
print(dna_filtered.shape)
print(mrna_filtered.shape)
print(mirna_filtered.shape)
```

```python
dna_filtered.head()
```

## **Data Integration**

```python
mrna_features = mrna_filtered.drop('sample ID', axis=1)
mirna_features = mirna_filtered.drop('sample ID', axis=1)
```

```python
omic_data = pd.concat([dna_filtered, mrna_features, mirna_features], axis=1)
omic_data.head()
```

```python
print(omic_data.shape)
```

```python
sampleid = omic_data['sample ID']
features = omic_data.drop(['sample ID'], axis=1)
```

```python
sampleid
```

```python
features
```

## **Data Normalization**

```python
from sklearn.preprocessing import MinMaxScaler
# Apply MinMax scaling to the features
scaler = MinMaxScaler(feature_range=(0, 1))
normalized_features = scaler.fit_transform(features)

# Reconstruct DataFrame with normalized data, keeping the original column names
normalized_data = pd.DataFrame(normalized_features, columns=features.columns)
```

```python
normalized_data.head()
```

```python
# Insert Sample ID column
normalized_data.insert(0, 'Sample ID', sampleid)
```

```python
normalized_data.head()
```

```python
# Save the normalized data to a new CSV file
normalized_data.to_csv(data_path + 'Normalized_omic.csv', index=False)

# Load and display the normalized data for verification
normalized_omics = pd.read_csv(data_path + 'Normalized_omic.csv')
print(normalized_omics)
```

```python
# Replace classification with values
label_values = label.replace(to_replace=["LumA","LumB","Basal","Her2","Normal"],value=["1","2","3","4","5"])
label_values
```

```python
# Insert Label column
omic_class = pd.concat([normalized_data, label_values], axis=1)
omic_class
```

```python
omic_class.to_csv(data_path + 'omic_class.csv', header=True, index=False)
```
